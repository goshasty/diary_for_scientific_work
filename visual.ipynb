```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from sklearn.cluster import DBSCAN
```

```python
def reader(file_name):
    list_obj = list()
    with open(file_name, 'r') as fd_file:
        i = 0;
        for line in fd_file:
            line = line.split()
            coords = np.zeros(shape=(3,))
            coords[0] = float(line[0])
            coords[1] = float(line[1])
            coords[2] = float(line[2])
            i += 1
            list_obj.append(coords)            
    return np.array(list_obj)
```

```python
mas = reader('r 0.05 0.075 xyz.txt')
```

```python
mas.shape
```

```python
%matplotlib notebook

max_num = 3000

#fig = plt.figure(figsize=(5, 5))
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')
ax.scatter(mas[:max_num,0], mas[:max_num,1], mas[:max_num,2], s=1)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
