```python
import numpy as np
```

```python
with open('log_1.txt') as fd:
    last_line = ''
    list_val = list()
    for i, line in enumerate(fd):
        if last_line[:12] == 'num clusters':
            list_val[-1].append(float(line))
            print(line)
        if line[:8] == 'expected':
            list_val.append(list())
        last_line = line
        
```

```python
not_null = list()
for b in list_val:
    if len(b) > 0:
        not_null.append(b)
```

```python
num_cells = 8

ap = np.zeros((5, num_cells))
for i, t in enumerate(not_null):
    ar = np.array(t)
    new_ar = ar[ar<1000000.]
    num = new_ar.shape[0]
    folds = num // num_cells 
    ap[i] += np.mean(new_ar[:folds*num_cells].reshape(-1, folds), axis=1)
    print((new_ar>10e5).sum())
    print(folds*num_cells, num)
```

```python
np.mean(ap, axis=0)
```

```python
for a in ap:
    print(a)
```
